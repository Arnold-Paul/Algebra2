```python
# converts string input (even fractions) to float
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans


# Simplest one-step addition
def one_step_add():
    import random
    # Display problem
    a = random.randint(-4,10)
    b = random.randint(2,24)
    print("x + ", a, " = ", b)
    ans = float(input("x = "))
    answer = b-a
    # Test input
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")


# One-step additon with negaive numbers
def one_step_subtract():
    import random
    a = random.randint(-19,-1)
    b = random.randint(2,24)
    print(a, " + x = ", b)
    ans = float(input("x = "))
    # test
    answer = b-a
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")

# One-step multiply
def one_step_mult():
    # Uses string_frac(<input string>)
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print(a, "x = ", b)
    print("Round your answer to two decimal places.")
    ans_in = (input("x = "))
    answer = round(b/a,2)
    # test
    if string_frac(ans_in)==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")


# One-step divide
def one_step_div():
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print("x/", a, " = ", b)
    ans = float(input("x = "))
    answer = b*a
    # test
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")


# Two-step problems
def two_step():
    import random
    # Uses string_frac()
    a = random.randint(1,11)
    b = random.randint(-7,12)
    c = random.randint(2,36)
    print(a, "x + ", b, " = ", c)
    print("Round answer to two decimal places")
    ans_in = input("x = ")
    answer = (c-b)/a
    # test
    if round(string_frac(ans_in),2)==round(answer,2):
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")


# test loop
for a in range(2):
    one_step_add()
    one_step_subtract()
    one_step_mult()
    one_step_div()
    two_step()
    print(" ")

two_step()
two_step()
```

```python
numerator = 12
denominator = 24
factor = 1

# Find greatest common factor
for test_factor in range(1,denominator+1):
    if numerator%test_factor==0 and denominator%test_factor==0:
        factor = test_factor

# Divide out greatest common factor
n = int(numerator/factor)
d = int(denominator/factor)

print("original: ", numerator, "/", denominator)
print("reduced: ", n, "/", d)
```

```python
from sympy import *

x,y = symbols('x y')

first = 2*x + y - 1
second = x - 2*y + 7


# parse finite set answer as coordinate pair
solution = linsolve([first, second], (x, y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

# Print a coordinate pair
print("(", x_solution, ",", y_solution, ")")
```

```python
from sympy import *
from sympy.plotting import plot
from sympy import sqrt

var('x y')

# First equation set equal to zero, ready to solve
first = -x**2 - y + 10
#first = sqrt(x) - y

#Second equation set equal to zero, ready to solve
second = 2*x**2 - 2*y - 4
#second = -x + 5 - y

# Solve - can be linear or nonlinear equations
solution = nonlinsolve([first, second], (x, y))
for a in range(len(solution.args)):
    x_solution = solution.args[a][0]
    y_solution = solution.args[a][1]
    print("Solution = (", x_solution, ",", y_solution, ")")

# Sympy syntax for equation equal to zero, ready to factor
y_first = Eq(first,0)

# Sympy solve for y
y1 = solve(y_first,y)

# Same two steps for second equation
y_second = Eq(second,0)
y2 = solve(y_second,y)

# Show factored results
print("y = ", y1[0])
print("y = ", y2[0])

# Plot solution
x = symbols('x')
xmin = -10
xmax = 10
plot(y1[0], y2[0], (x,xmin,xmax))

```

Graph a quadratic function

```python
import matplotlib.pyplot as plt
import numpy as np
import math

a = 1
b = 5
c = 6

print("y = ", a, "x**2 + ", b, "x + ", c)

# Vertex
vx = -b/(2*a)
vy = a*(vx**2) + b*vx + c
print("Vertex: (",vx,",",vy,")")

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 10*(xmax-xmin)
x = np.linspace(xmin,xmax,points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# Graph the parabola
y1 = a*x**2 + b*x + c
plt.plot(x, y1)

# Plot the vertex point
plt.plot([vx],[vy], 'ro')

# Find and plot the roots
d = b**2 - 4*a*c
if d>=0:
    root_1 = (-b + math.sqrt(d))/(2*a)
    root_2 = (-b - math.sqrt(d))/(2*a)
    plt.plot([root_1, root_2],[0,0], 'go')
    print("Roots: x = ", root_1, " and x = ", root_2)

plt.show()
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import math

# All graphing happens in this function
def f(a,b,c):
    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10
    points = 10*(xmax-xmin)
    x = np.linspace(xmin,xmax,points)

    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # x axis
    plt.plot([0,0],[ymin,ymax], 'b') # y axis

    # Parabola
    y1 = a*x**2 + b*x + c
    plt.plot(x, y1)

    # Vertex
    vx = -b/(2*a)
    vy = a*(vx**2) + b*vx + c
    plt.plot([vx],[vy], 'ro')

    # Roots
    d = b**2 - 4*a*c
    if d>=0:
        root_1 = (-b + math.sqrt(d))/(2*a)
        root_2 = (-b - math.sqrt(d))/(2*a)
        plt.plot([root_1, root_2],[0,0], 'go')

    # Set the equation as the title
    sa = str(a)
    sb = str(b)
    sc = str(c)
    h1 = "y = ", sa, "x**2 + ", sb, "x + ", sc
    h2 = ""
    for w in h1:
        h2 = h2 + w

    plt.title(h2)
    plt.show()

# Interactive creates sliders for a, b, and c values
interactive_plot = interactive(f, a=(1, 9), b=(-9,9), c=(-9,9))
interactive_plot

```

# Simple and Compound Interest

 Payment function without rounding

```python
def payment(p,r,t):
    numerator = (r/12)*(1+(r/12))**(12*t)
    denominator = (1+(r/12))**(12*t) - 1
    payment = p*numerator/denominator
    return payment

print(payment(200000,.06,30))
```

A good use of natural log: how long will it take for an investment to double in value?<br>
Start with the continuous growth formula:<br>
A = Pe<sup>rt</sup><br>
If the amount is double, then annuity is two times the principle:<br>
2P = Pe<sup>rt</sup><br>
Divide both sides by P and get this:<br>
2 = e<sup>rt</sup><br>
Because of the base <i>e</i>, take the natural log of both sides and get this:<br>
ln(2) = rt<br>
Then divide by r to solve for t.<br>
$\frac{ln(2)}{r}$ = t


```python
import math

# Rate of increase as a decimal
r = .09

# Natural log of 2 divided by rate
t = round(math.log(2)/r)

print("It will double in ", t, " years")
```

#Scatter Plot Game

This game will randomly generate points on a graph, and the player will have to input the (x, y) coordinates.

```python
import matplotlib.pyplot as plt
import numpy as np

def scatter_plot_game():
    # Generate random points
    num_points = 5
    x = np.random.randint(0, 20, num_points)
    y = np.random.randint(0, 20, num_points)

    plt.scatter(x, y, color='blue')
    plt.xlim(0, 20)
    plt.ylim(0, 20)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Scatter Plot Game')
    plt.grid(True)
    plt.show()

    correct_answers = 0
    for i in range(num_points):
        user_x = float(input(f"Enter x-coordinate for point {i+1}: "))
        user_y = float(input(f"Enter y-coordinate for point {i+1}: "))

        if user_x == x[i] and user_y == y[i]:
            print("Correct!")
            correct_answers += 1
        else:
            print(f"Incorrect. The correct coordinates were ({x[i]}, {y[i]}).")

    print(f"You got {correct_answers} out of {num_points} points correct.")

scatter_plot_game()
```

#Algebra Practice Game

This game will generate one-step and two-step algebra problems with random integer values, both positive and negative.

```python
import random

def algebra_practice_game():
    def one_step_problem():
        a = random.randint(1, 10)
        b = random.randint(-10, 10)
        print(f"Solve for x: {a}x = {b}")
        user_answer = float(input("Enter your answer: "))
        correct_answer = b / a
        if user_answer == correct_answer:
            print("Correct!")
        else:
            print(f"Incorrect. The correct answer is {correct_answer}")

    def two_step_problem():
        a = random.randint(1, 10)
        b = random.randint(-10, 10)
        c = random.randint(-10, 10)
        print(f"Solve for x: {a}x + {b} = {c}")
        user_answer = float(input("Enter your answer: "))
        correct_answer = (c - b) / a
        if user_answer == correct_answer:
            print("Correct!")
        else:
            print(f"Incorrect. The correct answer is {correct_answer}")

    while True:
        problem_type = random.choice(['one_step', 'two_step'])
        if problem_type == 'one_step':
            one_step_problem()
        else:
            two_step_problem()

        cont = input("Do you want to solve another problem? (yes/no): ").strip().lower()
        if cont != 'yes':
            break

algebra_practice_game()
```

#Projectile Game

This game displays a "wall" with random height and location. The player adjusts a parabolic path to clear the wall using sliders.

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import numpy as np

def projectile_game():
    fig, ax = plt.subplots()
    plt.subplots_adjust(left=0.1, bottom=0.3)

    # Generate wall
    wall_x = np.random.randint(5, 15)
    wall_height = np.random.randint(5, 10)
    ax.axvline(x=wall_x, color='r', linewidth=3)
    ax.axhline(y=wall_height, color='r', linewidth=3)

    # Sliders for adjusting a, b, c
    axcolor = 'lightgoldenrodyellow'
    ax_a = plt.axes([0.15, 0.2, 0.65, 0.03], facecolor=axcolor)
    ax_b = plt.axes([0.15, 0.15, 0.65, 0.03], facecolor=axcolor)
    ax_c = plt.axes([0.15, 0.1, 0.65, 0.03], facecolor=axcolor)

    s_a = Slider(ax_a, 'A', 0.1, 5.0, valinit=1)
    s_b = Slider(ax_b, 'B', -10.0, 10.0, valinit=0)
    s_c = Slider(ax_c, 'C', 0.0, 10.0, valinit=0)

    def update(val):
        ax.clear()
        ax.axvline(x=wall_x, color='r', linewidth=3)
        ax.axhline(y=wall_height, color='r', linewidth=3)

        a = s_a.val
        b = s_b.val
        c = s_c.val
        x = np.linspace(0, 20, 400)
        y = a * x**2 + b * x + c
        ax.plot(x, y, label=f'y = {a:.2f}x^2 + {b:.2f}x + {c:.2f}')
        ax.legend()
        plt.draw()

    s_a.on_changed(update)
    s_b.on_changed(update)
    s_c.on_changed(update)

    plt.show()

projectile_game()
```

#Financial Calculator

This calculator includes the following functions:

    Calculate annuity with monthly or continuous growth
    Calculate monthly mortgage payment
    Estimate retirement investment balance
    Determine how long until an amount doubles, given the rate
    Solve logarithmic equations
    Convert to (and from) scientific notation

```python
import math

def calculate_annuity(present_value, rate, periods, continuous=False):
    if continuous:
        return present_value * math.exp(rate * periods)
    else:
        return present_value * ((1 + rate)**periods - 1) / rate

def calculate_mortgage_payment(principal, annual_rate, years):
    monthly_rate = annual_rate / 12
    payments = years * 12
    return principal * (monthly_rate * (1 + monthly_rate)**payments) / ((1 + monthly_rate)**payments - 1)

def estimate_retirement_balance(initial_balance, monthly_contribution, annual_rate, years):
    monthly_rate = annual_rate / 12
    months = years * 12
    future_value = initial_balance * (1 + monthly_rate)**months
    for i in range(months):
        future_value += monthly_contribution * (1 + monthly_rate)**(months - i)
    return future_value

def time_to_double(rate):
    return math.log(2) / math.log(1 + rate)

def solve_logarithmic_equation(base, value):
    return math.log(value, base)

def to_scientific_notation(number):
    return f"{number:.2e}"

def from_scientific_notation(notation):
    return float(notation)

# Example usage
def main():
    while True:
        print("\nFinancial Calculator")
        print("1. Calculate Annuity")
        print("2. Calculate Monthly Mortgage Payment")
        print("3. Estimate Retirement Investment Balance")
        print("4. Determine Time to Double Amount")
        print("5. Solve Logarithmic Equations")
        print("6. Convert to Scientific Notation")
        print("7. Convert from Scientific Notation")
        print("8. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            present_value = float(input("Enter present value: "))
            rate = float(input("Enter growth rate (as a decimal): "))
            periods = int(input("Enter number of periods: "))
            continuous = input("Continuous growth? (yes/no): ").strip().lower() == 'yes'
            print("Annuity value:", calculate_annuity(present_value, rate, periods, continuous))

        elif choice == '2':
            principal = float(input("Enter principal amount: "))
            annual_rate = float(input("Enter annual interest rate (as a decimal): "))
            years = int(input("Enter number of years: "))
            print("Monthly mortgage payment:", calculate_mortgage_payment(principal, annual_rate, years))

        elif choice == '3':
            initial_balance = float(input("Enter initial balance: "))
            monthly_contribution = float(input("Enter monthly contribution: "))
            annual_rate = float(input("Enter annual interest rate (as a decimal): "))
            years = int(input("Enter number of years: "))
            print("Estimated retirement balance:", estimate_retirement_balance(initial_balance, monthly_contribution, annual_rate, years))

        elif choice == '4':
            rate = float(input("Enter growth rate (as a decimal): "))
            print("Time to double amount:", time_to_double(rate), "periods")

        elif choice == '5':
            base = float(input("Enter logarithm base: "))
            value = float(input("Enter value: "))
            print("Solution to logarithmic equation:", solve_logarithmic_equation(base, value))

        elif choice == '6':
            number = float(input("Enter number: "))
            print("Scientific notation:", to_scientific_notation(number))

        elif choice == '7':
            notation = input("Enter scientific notation: ")
            print("Decimal value:", from_scientific_notation(notation))

        elif choice == '8':
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

# Data graph explorer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO
import requests
from tkinter import Tk
from tkinter.filedialog import askopenfilename

# Function to load CSV from local file
def load_csv_from_local():
    Tk().withdraw()  # We don't want a full GUI, so keep the root window from appearing
    filename = askopenfilename()  # Show an "Open" dialog box and return the path to the selected file
    if filename:
        return pd.read_csv(filename)
    else:
        print("No file selected.")
        return None

# Function to load CSV from a URL
def load_csv_from_url(url):
    response = requests.get(url)
    if response.status_code == 200:
        return pd.read_csv(StringIO(response.text))
    else:
        print(f"Failed to fetch data. HTTP Status code: {response.status_code}")
        return None

# Load CSV from a predefined URL
def load_csv_from_predefined_url():
    url = "https://example.com/data.csv"  # Replace with your predefined URL
    return load_csv_from_url(url)

# Main function to handle CSV loading and plotting
def main():
    choice = input("Choose an option to load CSV:\n1. Local file\n2. URL input\n3. Predefined URL\nEnter 1, 2, or 3: ")
    if choice == '1':
        df = load_csv_from_local()
    elif choice == '2':
        url = input("Enter the URL: ")
        df = load_csv_from_url(url)
    elif choice == '3':
        df = load_csv_from_predefined_url()
    else:
        print("Invalid choice.")
        return

    if df is not None:
        # Print headings and first two rows
        print("Headings:", df.columns.tolist())
        print("First two rows:\n", df.head(2))

        # Store column names as a list
        columns = df.columns.tolist()

        # User input for columns to plot
        print("Available columns:", columns)
        col1 = input(f"Choose the first column from {columns}: ")
        col2 = input(f"Choose the second column from {columns} (leave empty for single column plot): ")

        if col1 not in columns or (col2 and col2 not in columns):
            print("Invalid column selection.")
            return

        # Convert to numpy arrays
        x = df[col1].to_numpy()
        if col2:
            y = df[col2].to_numpy()
            plt.scatter(x, y)  # Scatter plot for two columns
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.title(f"Scatter Plot of {col1} vs {col2}")
        else:
            plt.plot(x)  # Line plot for single column
            plt.xlabel('Index')
            plt.ylabel(col1)
            plt.title(f"Line Plot of {col1}")

        # Show plot
        plt.show()
    else:
        print("Failed to load data.")

if __name__ == "__main__":
    main()
```
